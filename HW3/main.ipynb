### TODO #4: 利用 Seaborn 進行 Heatmap 作圖，並進行說明
- 觀察每一個特徵與房價 (MEDV) 的關聯程度
- 使用 dr.corr() 來計算 feature 和 feature 之間的關係係數
- 數值愈高表示愈正相關，數值愈低表示愈負相關

````python
plt.figure(figsize=(16, 9))
# Please use a function of Pandas here.

# Plot a heatmap with Seaborn
sns.heatmap(
    # 輸入資料給熱點圖，自動計算相關係數
    df.corr(),                      
    # 顯示每個格子所代表的數字
    annot=True,                
    # 將格子形狀調整為正方形
    square=True
)
plt.show()
````

從熱點圖可以看出'RM'0.7與MEDV有最高的正相關, 'LASTA'-0.74是最高的負相關

其他像是'INDUS', 'NOX', 'TAX', 'PTRATIO'都有高於0.4的相關度

其餘的class相關度都低於0.4，所以在訓練時就沒有挑選

### TODO #5: 挑選特徵，並進行說明
- 觀察 heatmap 中具有"顯著"正相關或是"顯著"負相關的特徵
- 把這些特徵進行 concatenation，得到新的 dataframe

````python
X = pd.DataFrame(np.c_[df['INDUS'], df['NOX'], df['RM'], df['TAX'], df['PTRATIO'], df['LSTAT']], # Please put features inside the bracket
          columns = ['INDUS', 'NOX', 'RM', 'TAX', 'PTRATIO', 'LSTAT']# Please put feature names here 
        )
````

主要挑選與MEDV相關度在0.4以上的class來做為訓練資料

### TODO #6: 呈現特徵分佈，並進行說明
- 把你選取到的"重要特徵"進行subplots，畫出散佈圖 (scatter plots)
- 每一張圖的X軸是你選到的特徵、Y軸是房價('MEDV')

- INDUS : INDUS低MEDV較高，但INDUS=17附近也有較高的MEDV

- TAX : TAX在200~400間MEDV的分布差不多
- PTRATIO : 資料大多在右下，PTRATIO越高，MEDV可能越低
- LSTAT : 與MEDV有指數下降的關係，LSTAT一增加MEDV就掉很快

  也能從圖中看出RM及LSTAT與MEDV相關度較高

## 訓練模型前的準備
### TODO #7: 切分資料
- 利用 train_test_split，將資料切成 train set 和 test set
- 設定 train set 的比例為全部資料的 80%
- 請記得設定 random_state 以控制隨機過程

````python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, df['MEDV'], random_state=1012, train_size=0.8)

print(X_train.shape) #(404, 2)
print(X_test.shape) #(102, 2)
print(Y_train.shape) #(404, )
print(Y_test.shape) #(102, )
````

## 訓練模型並進行評估

### TODO #8: 訓練 Linear Regression 模型，並說明訓練結果
- 利用 scikit-learn，將資料 fit 在 LinearRegression
- 將結果繪製成散佈圖 (scatter plot)，其中 X軸 為 Y_pred、Y軸為 Y_true

````python
# Modeling
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, Y_train)
print("R2 : " + str(model.score(X_test, Y_test)))
print("MSE : " + str(MSE(Y_test, model.predict(X_test))))
````

````python
# plotting the y_test vs y_pred
plt.scatter(
    model.predict(X_test),
    Y_test,
    color='red'
)
plt.show()
````

Linear Regression訓練結果R2約為0.7, MSE約為30.3

準確率並不高，從圖表可以看出預測出來普遍偏高

### TODO #9: 訓練 Random Forest Regressor 模型
- 利用 scikit-learn，將資料 fit 在 RandomForestRegressor 模型
- 將結果繪製成散佈圖 (scatter plot)，其中 X軸 為 Y_pred、Y軸為 Y_true

````python
from sklearn.ensemble import RandomForestRegressor
````

````python
# Modeling
model = RandomForestRegressor(random_state=1)
model.fit(X_train, Y_train)
print(model)
print("R2 : " + str(model.score(X_test, Y_test)))
print("MSE : " + str(MSE(Y_test, model.predict(X_test))))
````

````python
# plotting the y_test vs y_pred
plt.scatter(
    model.predict(X_test),
    Y_test,
    color='red'
)
plt.show()
````

Random Forest Regressor訓練結果R2約為0.91, MSE約為9.22

準確率比Linear Regression高了不少，圖表顯示predict大部分都與true相近， 但還是有一些有點偏差

### TODO #10: 參數調整，並進行說明
- 請嘗試使用 Grid search 來進行調參
    - 可以自己寫 for-loop
    - 或是使用 sklearn.model_selection 中的 GridSearchCV 函式

````python
from sklearn.model_selection import GridSearchCV
````

````python
parameters = {'n_estimators': [100, 500, 1000], 'max_features' : ["auto", "sqrt", "log2"], 'max_depth': [None, 1, 2, 3], 'min_samples_split': [2, 3], 'random_state':[1]}
model = GridSearchCV(RandomForestRegressor(), parameters, cv=5, n_jobs=-1, verbose=1)
#model = RandomForestRegressor(n_estimators=1000, max_depth=80, min_samples_split=2, random_state=1)
model.fit(X_train, Y_train)
print(model.best_params_)
print("R2 : " + str(model.score(X_test, Y_test)))
print("MSE : " + str(MSE(Y_test, model.predict(X_test))))
````

````python
plt.scatter(
    model.predict(X_test),
    Y_test,
    color='red'
)
plt.show()
````

利用GridSearchCV找出的參數為max_depth: None, max_features: sqrt, min_samples_split: 2, n_estimators: 1000

與預設有差別的是node的split方式改為sqrt, tree的數量由100改為1000

R2上升到0.92，MSE降到7.88，圖表的部分差不多，但較少有預測偏離很多的

### TODO #11: 訓練 其他的 regression 模型，與上述兩種模型比較並說明訓練結果
- 將資料 fit 在其他的迴歸模型，可選擇 SVR、XGBRegressor 或其他...
- 利用 TODO #10 的工具來進行參數調整
- 將結果繪製成散佈圖 (scatter plot)，其中 X軸 為 Y_pred、Y軸為 Y_true

````python
from sklearn.svm import SVR
from xgboost.sklearn import XGBRegressor
````

````python
model_svr = SVR(kernel='rbf', C=1e3, gamma=0.1)
model_svr.fit(X_train, Y_train)
print("SVR:")
print("R2 : " + str(model_svr.score(X_test, Y_test)))
print("MSE : " + str(MSE(Y_test, model_svr.predict(X_test))))

xgb1 = XGBRegressor()
parameters = {'nthread':[4],
              'objective':['reg:linear'],
              'learning_rate': [.03, 0.05, .07],
              'max_depth': [5, 6, 7],
              'min_child_weight': [4],
              'silent': [1],
              'subsample': [0.7],
              'colsample_bytree': [0.7],
              'n_estimators': [500]}

model_xgb = GridSearchCV(xgb1,
                        parameters,
                        cv = 2,
                        n_jobs = 5,
                        verbose=True)

model_xgb.fit(X_train, Y_train)
print("XGBRegressor")
print("R2 : " + str(model_xgb.score(X_test, Y_test)))
print("MSE : " + str(MSE(Y_test, model_xgb.predict(X_test))))
````

````python
fig, ax = plt.subplots(1, 2, figsize=(10,5))
ax[0].scatter(
    model_svr.predict(X_test),
    Y_test,
    color='red'
)
ax[1].scatter(
    model_xgb.predict(X_test),
    Y_test,
    color='red'
)
ax[0].set_title('SVR')
ax[1].set_title('XGBRegressor')
plt.show()
````

使用了SVR與XGBRegressor

- SVR準確率大概只有5成，圖表顯示預測較分散

- XGBRegressor R2為0.92 MSE為7.5，使用GridSearchCV調整learning_rate及max_depth這兩個參數， 圖表顯示準確率的確很高，預測跟正解差不多
